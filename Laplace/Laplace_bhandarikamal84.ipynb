```python
#Problem1:Square in a box
import numpy as np
import matplotlib.pyplot as plt
import numpy.random as random

Nsteps = 10000
Ngrid = 100
phi = np.zeros((Ngrid, Ngrid))

for i in range(Ngrid):
    for j in range(Ngrid):
        # Define conducting square box at middle
        if 40 <= i <= 60 and 40 <= j <= 60:
            phi[i][j] = 1
        # Boundary condition
        elif i == 0 or j == 0 or i== Ngrid-1 or j==Ngrid-1:
            phi[i][j] = 0
        else:
            phi[i][j]= random.rand(1)
            
for k in range(Nsteps):
    for j in range(Ngrid):
        for i in range(Ngrid):
            if 40 <= i <= 60 and 40 <= j <= 60:
                phi[i][j] = 1
            elif 0 < i < Ngrid - 1 and 0 < j < Ngrid - 1:
                phi[i][j] = (phi[i - 1][j] + phi[i + 1][j] +
                               phi[i][j - 1] + phi[i][j + 1]) / 4
            else:
                phi[i][j] = 0
plt.imshow(phi)
plt.show()
```

```python
#Problem4:charge in a box
import numpy as np
import matplotlib.pyplot as plt
import numpy.random as random

Nsteps = 1000
Ngrid = 100
phi = np.zeros((Ngrid, Ngrid))

for i in range(Ngrid):
    for j in range(Ngrid):
        # Locate charge at centre
        if  i == 50 and  j == 50:
            phi[i][j] = 0
        # Boundary condition
        elif i == 0 or j == 0 or i== Ngrid-1 or j==Ngrid-1:
            phi[i][j] = 0
        else:
            phi[i][j]= 1/np.sqrt((i-50)**2 + (j-50)**2)
            
for k in range(Nsteps):
    for j in range(Ngrid):
        for i in range(Ngrid):
            if  i == 50 and  j == 50:
                phi[i][j] = 0
            elif 0 < i < Ngrid - 1 and 0 < j < Ngrid - 1:
                phi[i][j] = (phi[i - 1][j] + phi[i + 1][j] +
                               phi[i][j - 1] + phi[i][j + 1]) / 4
            else:
                phi[i][j] = 0
plt.imshow(phi)
plt.show()
```
